```python
%%writefile appm.py
import streamlit as st
import numpy as np
import joblib
import pickle

# -------------------------------
# Load your trained model
# -------------------------------
best_model = joblib.load("best_model.pkl")   # make sure file name matches

# -------------------------------
# Page Config
# -------------------------------
st.set_page_config(
    page_title="Forest Cover Prediction",
    layout="wide"
)

# -------------------------------
# Title
# -------------------------------
st.title("🌲 Forest Cover Type Prediction")
st.markdown("Predict forest cover type using environmental features")

# -------------------------------
# Sidebar Inputs
# -------------------------------
st.sidebar.header("🌿 Input Features")

def user_input():

    elevation = st.sidebar.slider("Elevation", 0, 4000, 2000)
    aspect = st.sidebar.slider("Aspect", 0, 360, 180)
    slope = st.sidebar.slider("Slope", 0, 60, 10)

    hor_dis_hydrology = st.sidebar.slider("Horizontal Distance to Hydrology", 0, 1000, 100)
    ver_dis_hydrology = st.sidebar.slider("Vertical Distance to Hydrology", -500, 500, 0)

    hor_dis_roadways = st.sidebar.slider("Horizontal Distance to Roadways", 0, 1000, 200)

    hillshade_9am = st.sidebar.slider("Hillshade 9am", 0, 255, 200)
    hillshade_noon = st.sidebar.slider("Hillshade Noon", 0, 255, 200)
    hillshade_3pm = st.sidebar.slider("Hillshade 3pm", 0, 255, 200)

    hor_dis_fire_points = st.sidebar.slider("Horizontal Distance to Fire Points", 0, 1000, 150)

    wilderness_area = st.sidebar.selectbox("Wilderness Area", [0, 1, 2, 3])
    soil_type = st.sidebar.selectbox("Soil Type", list(range(40)))

    data = {
        'elevation': elevation,
        'aspect': aspect,
        'slope': slope,
        'hor_dis_hydrology': hor_dis_hydrology,
        'ver_dis_hydrology': ver_dis_hydrology,
        'hor_dis_roadways': hor_dis_roadways,
        'hillshade_9am': hillshade_9am,
        'hillshade_noon': hillshade_noon,
        'hillshade_3pm': hillshade_3pm,
        'hor_dis_fire_points': hor_dis_fire_points,
        'wilderness_area': wilderness_area,
        'soil_type': soil_type
    }

    return pd.DataFrame([data])

input_df = user_input()

# -------------------------------
# Display Input
# -------------------------------
st.subheader("📊 Input Data")
st.write(input_df)

# -------------------------------
# Prediction
# -------------------------------
if st.button("🔍 Predict Cover Type"):

    try:
        prediction = best_model.predict(input_df)
        st.success(f"🌲 Predicted Cover Type: **{prediction[0]}**")

    except Exception as e:
        st.error("⚠️ Prediction failed. Check model features.")
        st.write(e)

# -------------------------------
# Footer
# -------------------------------
st.markdown("---")
st.markdown("Built with ❤️ using Streamlit")
```

```python
# Final code
import streamlit as st
import pandas as pd
import joblib

# ==============================
# Load Model + Encoder
# ==============================
best_model = joblib.load("best_model.pkl")
label_encoder = joblib.load("label_encoder.pkl")  # optional (if used)

# ==============================
# Page Config
# ==============================
st.set_page_config(page_title="Forest Cover Prediction", layout="wide")

st.title("🌲 Forest Cover Type Prediction App")
st.markdown("### Predict forest cover type using environmental features")

# ==============================
# Sidebar Inputs
# ==============================
st.sidebar.header("📥 Enter Feature Values")

# Numeric features
elevation = st.sidebar.number_input("Elevation", value=2000)
aspect = st.sidebar.number_input("Aspect", value=180)
slope = st.sidebar.number_input("Slope", value=10)

hor_dis_hydrology = st.sidebar.number_input("Horizontal Distance to Hydrology", value=100)
ver_dis_hydrology = st.sidebar.number_input("Vertical Distance to Hydrology", value=0)

hor_dis_roadways = st.sidebar.number_input("Horizontal Distance to Roadways", value=200)

hillshade_9am = st.sidebar.number_input("Hillshade 9am", value=200)
hillshade_noon = st.sidebar.number_input("Hillshade Noon", value=200)
hillshade_3pm = st.sidebar.number_input("Hillshade 3pm", value=200)

hor_dis_fire_points = st.sidebar.number_input("Horizontal Distance to Fire Points", value=150)

# Categorical (dropdown)
wilderness = st.sidebar.selectbox("Wilderness Area", [0, 1, 2, 3])
soil = st.sidebar.selectbox("Soil Type", list(range(40)))

# ==============================
# Create Input Data
# ==============================
data = {
    'elevation': elevation,
    'aspect': aspect,
    'slope': slope,
    'hor_dis_hydrology': hor_dis_hydrology,
    'ver_dis_hydrology': ver_dis_hydrology,
    'hor_dis_roadways': hor_dis_roadways,
    'hillshade_9am': hillshade_9am,
    'hillshade_noon': hillshade_noon,
    'hillshade_3pm': hillshade_3pm,
    'hor_dis_fire_points': hor_dis_fire_points
}

# One-hot encoding
for i in range(4):
    data[f'wilderness_area_{i}'] = 0

for i in range(40):
    data[f'soil_type_{i}'] = 0

# Set selected values
data[f'wilderness_area_{wilderness}'] = 1
data[f'soil_type_{soil}'] = 1

input_df = pd.DataFrame([data])

# ==============================
# Display Input
# ==============================
st.subheader("📊 Input Data")
st.write(input_df)

# ==============================
# Prediction
# ==============================
if st.button("🔍 Predict Cover Type"):
    try:
        pred = best_model.predict(input_df)

        # Inverse transform (if encoder used)
        try:
            pred_label = label_encoder.inverse_transform(pred)
            result = pred_label[0]
        except:
            result = pred[0]

        st.markdown("## 🎯 Prediction Result")
        st.success(f"🌲 Predicted Forest Cover Type: **{result}**")

    except Exception as e:
        st.error("⚠️ Prediction failed. Check model and features.")
        st.write(e)

# ==============================
# Footer
# ==============================
st.markdown("---")
st.markdown("Built with ❤️ using Streamlit | Final Year ML Project")
```
